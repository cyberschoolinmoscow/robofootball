```python
img = cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)

#print(img.shape)

plt.figure(figsize=(12, 11))

#color scheme (gray)
plt.imshow(img)
plt.show()

mask = cv2.inRange(img, (120, 130, 80), (230, 250, 200))

plt.figure(figsize=(12, 11))
plt.imshow(mask, cmap="gray")
plt.show()
```
